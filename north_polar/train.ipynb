## Import Dataset

```python
dir_path = 'model/dataset/polygons'
dataset = gee.importData(dir_path, enlarge=True)

print(f"Available countries are {list(dataset.keys())}.")
```

## Import Sentinel 2

```python
coords = dataset['Alaska']['coords']
bounds = dataset['Alaska']['bounds']

# coords = dataset['Canada']['coords']
# bounds = dataset['Canada']['bounds']
```

### 下载图像

## 批量下载

```python
for country in dataset.keys():
    coords = dataset[country]['coords']
    bounds = dataset[country]['bounds']

    pics_list = gee.exportBunchPics(bounds,
                                    start_date='2024-06-01',
                                    end_date='2024-08-31',
                                    vis_mode='urban_false_color')
    tmp = Downloader(gee, country=country)

    # tmp.download_from_pics_list(pics_list, output_size=(256, 256))
    tmp.download_bounds_list(pics_list)
    tmp.convert_pngs_to_geotiff()
```
